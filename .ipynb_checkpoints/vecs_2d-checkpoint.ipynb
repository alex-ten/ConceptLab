```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
%matplotlib notebook

xylim = 20

rbs = widgets.RadioButtons(options={'v': 0, 'w': 1}, description='vector: ')
# v_check = widgets.Checkbox(value=False, description='vector v: ')
# w_check = widgets.Checkbox(value=False, description='vector v: ')

x = widgets.FloatSlider(value=1, min=-xylim, max=xylim, description='x = ')
y = widgets.FloatSlider(value=1, min=-xylim, max=xylim, description='y = ')

c = widgets.Textarea(value='1', placeholder='scalar for v',description='c = ')
d = widgets.Textarea(value='1', placeholder='scalar for w',description='d = ')

plt.ioff()

fig = plt.figure()
ax = fig.add_subplot(111)

v = ax.annotate("", xy=(float(c.value)*x.value, 0), xytext=(0, 0), arrowprops=dict(arrowstyle="->", linewidth=3, color='green'))
w = ax.annotate("", xy=(0, float(d.value)*y.value), xytext=(0, 0), arrowprops=dict(arrowstyle="->", linewidth=3, color='blue' ))

ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['left'].set_smart_bounds(True)
ax.spines['bottom'].set_smart_bounds(True)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# ax.set_xlim = (-xylim, xylim)
# ax.set_ylim = (-xylim, xylim)
ax.set_aspect('equal')
# ax.axhline(y=0, color='k')
# ax.axvline(x=0, color='k')
# ax.grid(True, which='both')




plt.ion()
plt.show()

vecs = (v,w)
cols = ('green','blue')


# def redraw_arrow(arrow, new_x, new_w, col):
#     ax.clear()
#     ax.arrow(0,0, new_x, new_y, color=col, linewidth=2)
    
# widgets.interact(redraw_arrow, arrow=vecs[rbs.value], new_x=x, new_y=y, color=cols[rbs.value])
```
